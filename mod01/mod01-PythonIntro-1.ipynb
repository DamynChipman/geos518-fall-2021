```python
import numpy as np
import matplotlib.pyplot as plt

Nt = 200
mu_x = 0.75
s_x = 0.75
```

```python
x_test = np.zeros(Nt)

for i in np.arange(Nt):

    x_i = s_x*np.random.randn() + mu_x
    
    if(i==0):
        x_test[i] = 0.0 + x_i
    else:
        x_test[i] = x_test[i-1] + x_i
```

```python
plt.rcParams.update({'font.size': 16})

plt.figure(figsize=(16,10))
plt.plot(np.arange(Nt),x_test,'b-')
plt.xlabel('Number of Steps')
plt.ylabel('Distance Travelled [m]')
```

```python
Nx = 1000

X = np.zeros((Nt,Nx))

for i in np.arange(Nt):

    for j in np.arange(Nx):
        x_i = s_x*np.random.randn() + mu_x
    
        if(i==0):
            X[i,j] = 0.0 + x_i
        else:
            X[i,j] = X[i-1,j] + x_i
```

```python
plt.figure(figsize=(16,10))
plt.plot(np.arange(Nt),X,'b-')
plt.xlabel('Number of Steps')
plt.ylabel('Distance Traveled [m]')
```

```python
mean_X = np.mean(X,axis=1)
std_X = np.std(X,axis=1)

plt.figure(figsize=(16,20))
plt.subplot(2,1,1)
plt.plot(np.arange(Nt),mean_X,'k-')
plt.ylabel('Average Distance Traveled [m]')

plt.subplot(2,1,2)
plt.plot(np.arange(Nt),std_X,'k-')
plt.xlabel('Number of Steps')
plt.ylabel('Standard Deviation Distance Traveled [m]')
```
